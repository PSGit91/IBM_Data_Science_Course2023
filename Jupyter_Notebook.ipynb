### **The following code compute the exponential moving average of a data series**

```python
import pandas as pd
import matplotlib.pyplot as plt

value = pd.DataFrame({'Values':[10,100,90,70,65,57,80,85,101,110,111,112,130,170,178,158,150,135,180,182]})
EMA   = value.ewm(com=0.5).mean()
                     
# Create plot

plt.plot(value.Values, label="Values")
plt.plot(EMA, label="EMA Values")
plt.legend()
plt.show()
```

- import pandas and matplotlib libraries
- create a Pandas dataframe or import from file
- call Exponential Moving Average function (ewm)
- plot the results
